```python
df = pd.DataFrame({"x": [1, 2]}).assign(label="あああ")
g = (
    p9jp.ggplot(df, p9.aes(x="x", y="x", label="label"))
    + p9.geom_text()
    + p9.labs(x="あああ", y="亜々ゝ", title="アアア")
    + p9jp.theme_classic()
)
g.show()
```
